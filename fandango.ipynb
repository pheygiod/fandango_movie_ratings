# Fandango & Chill: Investigating Biased Movie Ratings

# Introduction

The purpose of this project is to test the reliability of Fandango's movie rating system. Fandango is an online movie ratings aggregator. It displays a 5-star rating system on their website from 0 to 5. In 2015, Walt Hickey, a data journalist, found discrepancies between the number of stars and the actual rating. After Hickey published his article, Fandango's officials replied apologising. They explained that a bug in their system caused the biased rounding off. They have presumably fixed the error. Yet, the actual rating values are not displayed anymore in the pages' HTML. This means that we cannot tell for sure.

We will now check if there has been any change after Hickey's analysis. We will use two sources of data: the data Hickey collected, and a dataset of Fandango's movie ratings. The latter was collected after Hickey's published his article.

There's a slight difference between Fandango's ratings for movies in 2015 and Fandango's ratings for movies in 2016. Although this might be because Fandango's officials fixing the biased rating system, we have no tangible proof of this hypothesis. Indeed, the rating values are not displayed anymore in Fandango's pages' HTML.

# Sampling

Now that we've explored the two datasets, it's time to create our samples. The goal is to check if there's been any change in Fandango's rating system since Hickey's article. For now, all movie ratings on Fandango's website will be our population of interest.

Let's sample our populations at two different time periods: movie ratings before and after Hickey's article. We need our samples to be as representative of the population as possible.

## Is our Sample Representative?

For the dataset before Hickey's analysis we have a sample of 146 films. The sampling criteria used are:

- the film must have had at least 30 fan reviews on the Fandango's website before Aug. 24, 2015, and

- the movie must have had tickets on sale in 2015.

Such criteria do not make the sample random. This is because all those movies that do not meet the criteria were not included. Thus, this sampling is not representative of the entire population.

As for the second dataset, we have movie rating data for 214 of the most popular movies. The sampling criteria are:

- the movie must have had a significant number of votes and reviews.

- the movie must have been released in 2016 and 2017.

- the movie must have had reviews on Fandango as of March 22, 2017.

This dataset does not have a random sample either. Any movie that does not meet the criteria will be excluded from the sample. So we can state that these two samples might have been useful for their analysts' research. Yet, we will not consider them helpful for our analysis. It is therefore necessary for us to change the goal of our analysis by placing some limitations to it. In this way, the samples will become more representative.

# Comparing Distribution Shapes for 2015 and 2016

Now we can start analysing the two isolated samples. Let's make a comparison between the shapes of the distributions of movie ratings. ## Determining the Direction of the Change

We'll take a couple of summary statistics to get a more precise picture about the direction. We'll take each distribution of movie ratings and compute its mean, median, and mode. Then, we'll compare these statistics to determine what they tell about the direction.

```python
# Compute means
mean_2015 = fandango_2015_df['fandango_stars'].mean()
mean_2016 = fandango_2016_df['fandango'].mean()

# Compute medians
median_2015 = fandango_2015_df['fandango_stars'].median()
median_2016 = fandango_2016_df['fandango'].median()

# Compute modes
mode_2015 = fandango_2015_df['fandango_stars'].mode()[0]
mode_2016 = fandango_2016_df['fandango'].mode()[0]

# Create dataframe of computed statistics 
summary = pd.DataFrame()
summary['2015'] = [mean_2015, median_2015, mode_2015]
summary['2016'] = [mean_2016, median_2016, mode_2016]
summary.index = ['mean', 'median', 'mode']
summary.round(1)
```

```python
# Generate a grouped bar plot
plt.style.use('fivethirtyeight')
summary['2015'].plot.bar(color = '#a98d19', align = 'center', label = '2015', width = .25)
summary['2016'].plot.bar(color = 'purple', align = 'edge', label = '2016', width = .25,
                         rot = 0, figsize = (8,5))

plt.title('Comparing summary statistics: 2015 vs 2016', y = 1.06)
plt.ylim(0,5.5)
plt.yticks(arange(0,5.1,.5))
plt.ylabel('Stars')
plt.legend(framealpha = 0, loc = 'upper center')
plt.show()
```

The mean of movie ratings in 2016 is lower than the mean in 2015 by 0.2. This shows a drop of almost 5% of the mean of movie ratings in 2016 relative to the mean ratings in 2015. There's also no difference between the median of movie ratings in 2015 and the median in 2016. Moreover, the most frequent star movie rating in 2016 was 4.0. This is lower compared to the most popular movie rating of 2015, that is, 4.5. So on average, fans who rated movies in 2016 were more likely to rate them lower compared to those who rated the movies in 2015.
